# Processamento de imagem para corrigir as amostras de texto em uma imagem.

### Para o processamento de imagem foi utilizado o Spacy
Scipy e numpy trabalham juntos para lidar com interpolação, integração, otimização, processamento de imagem,

```python
from scipy import signal
from PIL import Image
import numpy as np
import cv2
import matplotlib.pyplot as plt
import math
from scipy import ndimage
%matplotlib inline
```

```python
def load_image(path):
    """ Carrega a imagem """
    return np.asarray(Image.open(path))/255.0

def denoise_image(image):
    """ Use um filtro passa-baixo com um kernel de 9 
        para borrar a imagem e reduzir as características da fonte para obter uma imagem de fundo """
    
    bg = signal.medfilt2d(image, 9) # A matriz bidimensional do filtro mediano. o Kernel deve ser um numero impar
    mask = image < bg - 0.1
    
    # Após a saída dos valores de mask, inp e bg, pode-se descobrir que no intervalo de 0-1, 
    # quanto mais próximo de 0, mais escura é a cor. Portanto, o valor do vetor em bg é maior que o valor em inp. 
    # A parte branca em inp está próxima de 1 e a parte preta está próxima de 0. Portanto, mask = inp <bg -0.1 
    # A instrução obterá uma matriz binária cujo valor é apenas 1 ou 0

    # Se a máscara for a parte do fundo (False), retorne 1. 
    # Se for a parte da fonte (True), 
    # insira o valor da parte da fonte correspondente em inp.
    return np.where(mask, image, 1.0)

def rotate_image(image):
    '''
    Função para corrigir a rotação de documentos.
    
    params:
        image : Imagem do documento.
        
    return:
        img_rotated : Imagem rotacionada
    
    '''
    img_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    img_edges = cv2.Canny(img_gray, 200, 200, apertureSize=3)
    lines = cv2.HoughLinesP(img_edges, 1, math.pi / 180.0, 100, minLineLength=100, maxLineGap=5)

    angles = []

    for [[x1, y1, x2, y2]] in lines:
        angle = math.degrees(math.atan2(y2 - y1, x2 - x1))
        angles.append(angle)

    median_angle = np.median(angles)
    img_rotated = ndimage.rotate(img_gray, median_angle)

    thresh = 160
    im_bw = cv2.threshold(img_rotated, thresh, 255, cv2.THRESH_BINARY)[1]
    
    return im_bw
```

## Testando as funções de tirar o ruido e rotacionar a imagem

```python
inp_path = 'noisy_data/155.png'
out_path = 'output.png'

image = load_image(inp_path)
image_not_denoise = denoise_image(image)

tmp = np.asarray(image_not_denoise*255.0, dtype=np.uint8)
tmp = Image.fromarray(tmp)
im_bw = rotate_image(np.array(tmp.convert('RGB')))
```

```python
plt.rcParams['figure.figsize'] = (20,30)
plt.imshow(im_bw)
```

```python
# lendo todos os arquivos da pasta aplicando transformacao e salvando em outra pasta.
directory = 'noisy_data/'

files_path = [os.path.abspath(os.path.join(directory, p)) for p in os.listdir(directory)]
for file in files_path:
    image = load_image(file)
    image_not_denoise = denoise_image(image)
    tmp = np.asarray(image_not_denoise*255.0, dtype=np.uint8)
    tmp = Image.fromarray(tmp)
    im_bw = rotate_image(np.array(tmp.convert('RGB')))
    cv2.imwrite(newpath + "/" + file.split('\\')[-1], im_bw) 
```
